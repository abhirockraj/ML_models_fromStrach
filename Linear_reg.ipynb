```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
c= np.ones(10,dtype=int)
a = np.random.randint(1,10,10,dtype=int)
b = np.random.randint(1,10,10,dtype=int)
```

```python
print (c)
print(a)
print(b)
```

```python
x= np.vstack((c,a,b))
x.shape
```

```python
th = np.array([5,2,3])
th=np.reshape(th,(1,3))
th.shape
```

```python
# y = 5*c+2*a+3*b
```

```python
y = np.dot(th,x)
y
```

```python
th1 = np.zeros((1,3), dtype=int)
th1
```

```python
cost =[]
for i in range(0,1000):
    yhat = np.matmul(th1,x)
    loss = yhat-y
    cost.append((1/(2*3))*(np.sum(np.square(loss))))
    th1 = th1- ((0.001/3)*(np.matmul(loss,np.transpose(x))))
   
#     print(th1)
    
```

```python
l=np.arange(1,1001)
# print(l)
```

```python
plt.plot(l, cost, label = "line 2")
  
# naming the x axis
plt.xlabel('iteration')
# naming the y axis
plt.ylabel('cost')
# giving a title to my graph
plt.title('Cost function LR!')
  
# show a legend on the plot
plt.legend()
  
# function to show the plot
plt.show()
```

```python
ypredict= np.reshape(ypredict,(10,1))
y = np.reshape(y,(10,1))
```

```python
l1 = np.arange(1,11)
plt.plot(l1, y, color='r',label = "actual")
plt.plot(l1,ypredict,color='g',label="predicted" ) 
# naming the x axis
plt.xlabel('iteration')
# naming the y axis
plt.ylabel('traget')
# giving a title to my graph
plt.title('predicted vs actuall')
  
# show a legend on the plot
plt.legend()
  
# function to show the plot
plt.show()
```
